```python
import numpy as np
import scipy as sp
import pandas as pd
import sqlite3
```

```python
##Sqlite path varaible -- change for appropriate simulation
#Connect to .sqlite data output file
sqlite_file_path = "C:\\Users\\am122\\Desktop\\Wireless Visualization Scripts\\ConferenceRoom.FilteringAoATest2.sqlite"
conn = sqlite3.connect(sqlite_file_path)

#Output folder path
output_folder = 'C:\\Users\\am122\\Desktop\\Wireless Visualization Scripts\\TopKTest\\'
```

```python
#View database tables
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

print("Data tables:")
for table in tables:
    print(table[0])
```

```python
#Number of antennas in MIMO array
tx_num_antennas = 4
rx_num_antennas = 4

#Select top K paths
K = 5
```

```python
combined_top_k = pd.DataFrame()

for tx in range(0,tx_num_antennas):
    for rx in range(0,rx_num_antennas):
        desired_fields = '*'
        query = f"SELECT {desired_fields} FROM path_utd WHERE tx_sub_antenna = {tx} AND rx_sub_antenna = {rx};"
        sub_paths = pd.read_sql_query(query, conn)
        top_k = sub_paths.nlargest(K, 'received_power')
        combined_top_k = pd.concat([combined_top_k, top_k])
combined_top_k
```

```python
table_name = 'tx'
query = f"SELECT * FROM {table_name};" 
tx_locations = pd.read_sql_query(query, conn)
tx_locations
```

```python
table_name = 'rx'
query = f"SELECT * FROM {table_name};" 
rx_locations = pd.read_sql_query(query, conn)
rx_locations
```

```python
combined_top_k.to_csv(f"{output_folder}top_{K}_path_information.csv", index=False)
rx_locations.to_csv(f"{output_folder}rx_locations.csv", index=False)
tx_locations.to_csv(f"{output_folder}tx_locations.csv", index=False)
```
